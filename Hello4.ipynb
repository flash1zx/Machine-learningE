```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.species.value_counts()
```

```python
sns.set_theme(style="dark")
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", data=iris, hue="species", palette="rainbow_r")
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
sns.barplot(x="sex", y="survived", data=titanic, hue="class", palette="RdBu")
```

```python
sns.countplot(x="embarked", data=titanic)
```

```python
sns.boxplot(x="species", y="sepal_width", data=iris)
```
