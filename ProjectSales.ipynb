Question:1
Which month has the highest amount of sales in the dataset??

```python
dataset.head()
```

```python
dataset.Product.value_counts()
```

Removing Product Column

```python
dataset = dataset[dataset["Product"] != "Product"]
```

```python
dataset.shape
```

```python
dataset.dtypes
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset["Quantity Ordered"] = pd.to_numeric(dataset["Quantity Ordered"])
```

```python
dataset["Price Each"] = pd.to_numeric(dataset["Price Each"])
```

```python
dataset.info()
```

```python
dataset["Sales"] = dataset["Price Each"] * dataset["Quantity Ordered"]
```

```python
dataset.head
```

```python
dataset["Order Date"].str[:2]
```

```python
dataset["Month"] = dataset["Order Date"].str[:2]
```

```python
dataset.head()
```

```python
dataset.Month.value_counts()
```

```python
dataset.groupby("Month")["Sales"].sum().sort_values(ascending=False)
```

Answer to question 1:
Therfore, 12th month which is december is the month with the highest amount of Sales
